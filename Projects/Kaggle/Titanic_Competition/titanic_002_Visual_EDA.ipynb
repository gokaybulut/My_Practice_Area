```python
# Visual Exploratory Data Analysis
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt

# read training data to a DataFrame df
df = pd.read_csv('train_titanic.csv')

# get numerical columns
num_cols = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']

# show boxplots
for col in num_cols:
    df.plot(y = col, kind = 'box')
    plt.show()
    
# show scatterplots
#for col in num_cols:
#    df.plot(x = 'Survived', y = col, kind = 'scatter')
#    plt.show()

num_df = df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']]
_ = pd.plotting.scatter_matrix(num_df, figsize = [8,8], s=150, marker='D')
plt.show()
```
